```python
# 1. Clean age
def age_feature_engineer(df):
    '''
    Replace outlier ages with -999 and correct age if birth year is put in.
    '''
    df = df.fillna('NA')
    df.age = df.age.replace("NA","-999").astype(int)
    df.loc[df.age>=2004,'age'] = -999
    df.loc[(df.age>100) & (df.age<200),'age'] = 100
    df.loc[(df.age>200) & (df.age<2004),'age'] = df["date_account_created"].apply(lambda x: 
                                            datetime.strptime(x, "%Y-%m-%d")).dt.year - df.age
    return df
```

```python
# 2. Clean time variables
def time_var_feature_engineer(df):
    ''' 
    Create two new features - 
        1. date_diff_ac_fa (date differences between active date and account created)
        2. first_active_hour (first active hour of the day )
    '''
    df['first_active_date'] = df['timestamp_first_active'].astype(str).apply(lambda x: \
                                                                                      datetime.strptime(x[:8], "%Y%m%d"))
    df['account_created_date'] = df['date_account_created'].apply(lambda x: datetime.strptime(x, "%Y-%m-%d"))
    df['first_active_hour'] = df['timestamp_first_active'].astype(str).apply( \
                                    lambda x: x[8:10]).astype(int)
    df['date_diff_ac_fa'] = df['account_created_date'] - df['first_active_date']
    df['date_diff_ac_fa'] = df['date_diff_ac_fa'].dt.days
    df['first_active_year'] = df['first_active_date'].dt.year
    df['first_active_month'] = df['first_active_date'].dt.month
    df = df.drop(['timestamp_first_active', 'date_account_created', 'first_active_date', 'account_created_date'],axis=1)
    return df
```

```python
def data_preprocessing(df,train=True):
    '''
    Data preprocessing pipeline
    '''
    df_processed = df.drop(["date_first_booking"], axis=1)
    df_processed = age_feature_engineer(df_processed)
    df_processed = time_var_feature_engineer(df_processed)
    df_processed = df_processed.merge(df_session_merged,left_on='id',right_on='user_id',how='inner')
    
    if train == True:
        y_final = df_processed["country_destination"]
        df_processed = df_processed.drop(['country_destination'],axis=1)
        global onehotencode 
        onehotencode = OneHotEncoder(handle_unknown='ignore')
        df_onehot = onehotencode.fit_transform(df_processed.select_dtypes(exclude=np.number).drop('id',axis=1))
    else:
        y_final = "Hello world."
        df_onehot = onehotencode.transform(df_processed.select_dtypes(exclude=np.number).drop('id',axis=1))
    df_onehot = pd.DataFrame.sparse.from_spmatrix(df_onehot)
    df_numeric = df_train_processed.select_dtypes(include=np.number)
    df_final = pd.concat([df_numeric,df_onehot],axis=1)
    
    return df_final,y_final
```

### 5.2 Prepare custom transformers for pipeline 

```python
# 1. Create new features - title, family_size, isalone, cabin_level
class CreateNewFeature(BaseEstimator, TransformerMixin):
    def __init__(self):
        return None
    def fit(self,X,y=None):
        return self
    def transform(self,X,y=None):
        # Extract title
        X_new = X.copy()
        X_new['title'] = X_new['Name'].str.extract(r'(, .*\.)',expand=True)
        X_new['title'] = X_new['title'].str.replace(',','').str.strip()
        X_new.loc[X_new['title']=='Ms.', 'title'] = 'Miss.'
        X_new.loc[X_new['title']=='Mrs. Martin (Elizabeth L.', 'title'] = 'Mrs.'
        X_new.loc[~X_new['title'].isin(['Mr.', 'Mrs.', 'Miss.', 'Master.']), 'title'] = "Misc"
        # Create family_size and isalone
        X_new['family_size'] = X_new['SibSp'] + X_new['Parch'] + 1
        X_new['isalone'] = 0
        X_new.loc[X_new['family_size']>1, 'isalone'] = 1
        # Extract cabin_level from Cabin
        X_new['cabin_level'] = X_new['Cabin'].str.slice(start=0,stop=1)
        X_new.loc[X_new['cabin_level']=="N",'cabin_level'] = "None"
        return X_new
# 2. Drop irrelevant variables ('PassengerId', 'Ticket', 'Cabin', 'family_size', 'Name')
class DropVariables(BaseEstimator, TransformerMixin):
    def __init__(self):
        return None
    def fit(self,X,y=None):
        return self
    def transform(self,X,y=None):
        X_new = X.drop(['PassengerId', 'Ticket', 'Cabin', 'Name'],axis=1)
        # , 'family_size'
        return X_new
# 3. Fill in missing values using median or large number(100000)
class MissingImputer(BaseEstimator, TransformerMixin):
    def __init__(self, large_value=False, strategy="median"):
        self.strategy=strategy
        self.large_value=large_value
        self.missinglist = ['Age','Fare','Embarked','cabin_level']
        return None
    def fit(self,X,y=None):
        self.mode_impute = SimpleImputer(strategy='most_frequent')
        self.mode_impute.fit(X[['Embarked']])
        if self.large_value:
            return self
        else:
            self.miss_impute = SimpleImputer(strategy=self.strategy)
            self.miss_impute.fit(X[['Fare','Age']])
            return self
    def transform(self, X, y=None):
        X_new = X.copy()
        X_new.loc[:,'Embarked'] = self.mode_impute.transform(X[['Embarked']])
        X_new['cabin_level'] = X['cabin_level'].fillna('None')
        if self.large_value:
            X_new['Fare'] = X['Fare'].fillna(100000)
            X_new['Age'] = X['Age'].fillna(100000)
        else:
            X_new.loc[:,['Fare','Age']] = self.miss_impute.transform(X[['Fare','Age']])
        return X_new
# 4. Create dummies
class CategoricalDummies(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.categoricaldummy_list = ['Pclass','Sex','Embarked','title','isalone','cabin_level']
        # ,'SibSp','Parch','cabin_level'
        return None
    def fit(self,X,y=None):
        self.categorical_dummies = OneHotEncoder(handle_unknown='ignore')
        self.categorical_dummies.fit(X[self.categoricaldummy_list])
        return self
    def transform(self, X, y=None):
        X_new = self.categorical_dummies.transform(X[self.categoricaldummy_list])
        feature_names = self.categorical_dummies.get_feature_names(self.categoricaldummy_list)
        X_new_dummies = pd.DataFrame.sparse.from_spmatrix(X_new, columns=feature_names, index=X.index)
        X_new = X.drop(self.categoricaldummy_list,axis=1).join(X_new_dummies)
        return X_new
'''
    # 7. Standard Scaler - Return a dataframe
class StandardScalerDF(BaseEstimator, TransformerMixin):
    def __init__(self):
        return None
    def fit(self,X,y=None):
        global std_scaler
        std_scaler = StandardScaler()
        std_scaler.fit(X)
        return self
    def transform(self, X, y=None):
        X_new = X.copy()
        X_new = std_scaler.transform(X)
        X_new = pd.DataFrame(X_new, index=X.index, columns=X.columns)
        return X_new
'''
```

```python
# Preprocessor pipeline setup
preprocessor = Pipeline([('create_new_feature', CreateNewFeature()),('drop_variable',DropVariables()) , 
                         ('missing_imputer', MissingImputer(large_value=True)), ('create_dummies',CategoricalDummies())])
X_train_preprocessed = preprocessor.fit_transform(X_train)
X_validation_preprocessed = preprocessor.transform(X_validation)
X_fullset_preprocessed = preprocessor.transform(X_fullset)
X_predict_preprocessed = preprocessor.transform(X_predict)
column_final = X_train_preprocessed.columns
```
